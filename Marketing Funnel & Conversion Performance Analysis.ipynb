# TASK 3 - Marketing Funnel & Conversion Performance Analysis
# Dataset : Marketing Funnel by Olist (Kaggle)
# Tool    : Python + Jupyter Notebook

```python
# 1. Import Necessary Libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print(" Libraries imported successfully!")

# 2. Load the Datasets 
# Load the top-of-funnel leads (people who clicked or showed interest)
leads = pd.read_csv("olist_marketing_qualified_leads_dataset.csv")

# Load the bottom-of-funnel deals (people who actually converted/bought)
deals = pd.read_csv("olist_closed_deals_dataset.csv")

print(f"\n Total Leads Generated: {leads.shape[0]:,}")
print(f" Total Closed Deals: {deals.shape[0]:,}")

# 3. Merge the Data to Build the Funnel 
# We use a 'left' merge to attach the closed deals to our main leads table.
# If a lead didn't close a deal, those new columns will just be blank (NaN).
funnel = pd.merge(leads, deals, on='mql_id', how='left')

# Create a simple 1 or 0 column so we can easily calculate math later:
# 1 = They bought something (won_date is not blank)
# 0 = They dropped off (won_date is blank)
funnel['converted'] = funnel['won_date'].notna().astype(int)

print("\n Datasets successfully merged into the Funnel!")
print(f" Overall Funnel Conversion Rate: {(funnel['converted'].mean() * 100):.2f}%")

# 4. Inspect the Data 
# Let's peek at the first 3 rows to see our new table
display(funnel.head(3))
```

# SECTION 2 - Funnel Drop-off & Channel Analysis

```python
# 1. Identify Conversion Drop-offs 
total_leads = len(funnel)
total_conversions = funnel['converted'].sum()
total_dropoffs = total_leads - total_conversions

print(" OVERALL FUNNEL PERFORMANCE:")
print(f"1. Total Leads (Top of Funnel): {total_leads:,}")
print(f"2. Closed Deals (Bottom of Funnel): {total_conversions:,}")
print(f"3. Drop-offs: {total_dropoffs:,} leads lost in the funnel.")
print(f" Average Conversion Rate: {(total_conversions / total_leads * 100):.2f}%\n")

# 2. Channel Performance Analysis
print(" PERFORMANCE BY MARKETING CHANNEL:")

# Group the data by 'origin' (the marketing channel where the lead came from)
channel_stats = funnel.groupby('origin').agg(
    Total_Leads=('mql_id', 'count'),
    Converted_Deals=('converted', 'sum')
).reset_index()

# Calculate the conversion rate for each specific channel
channel_stats['Conversion_Rate (%)'] = (channel_stats['Converted_Deals'] / channel_stats['Total_Leads']) * 100

# Sort the table so the channels that brought in the most leads are at the top
channel_stats = channel_stats.sort_values(by='Total_Leads', ascending=False)

# Make the table look incredibly clean and highlight the best rates in blue
display(channel_stats.style.format({
    'Total_Leads': '{:,}',
    'Converted_Deals': '{:,}',
    'Conversion_Rate (%)': '{:.2f}%'
}).background_gradient(subset=['Conversion_Rate (%)'], cmap='Blues'))
```

# SECTION 3 - Export Clean Data for Power BI

```python
# Save our merged funnel to a new CSV file
funnel.to_csv('Cleaned_Olist_Funnel.csv', index=False)

print("Clean funnel dataset successfully exported as 'Cleaned_Olist_Funnel.csv'!")
print("Ready for Power BI!")
```
